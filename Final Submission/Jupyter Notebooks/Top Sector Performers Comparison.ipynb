```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime
from datetime import datetime, timedelta

aapl_df = pd.read_csv("AAPL.csv")
amzn_df = pd.read_csv("AMZN.csv")
brk_df  = pd.read_csv("BRK-B.csv")
hon_df  = pd.read_csv("HON.csv")
jnj_df  = pd.read_csv("JNJ.csv")
lin_df  = pd.read_csv("LIN.csv")
nee_df  = pd.read_csv("NEE.csv")
pg_df   = pd.read_csv("PG.csv")
xom_df  = pd.read_csv("XOM.csv")

aapl_df["aapl_open"] = aapl_df["Open"].astype("float64")
aapl_df = aapl_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

amzn_df["amzn_open"] = amzn_df["Open"].astype("float64")
amzn_df = amzn_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

brk_df["brk_open"] = brk_df["Open"].astype("float64")
brk_df = brk_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

hon_df["hon_open"] = hon_df["Open"].astype("float64")
hon_df = hon_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

jnj_df["jnj_open"]  = jnj_df["Open"].astype("float64")
jnj_df = jnj_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

lin_df["lin_open"] = lin_df["Open"].astype("float64")
lin_df = lin_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

nee_df["nee_open"] = nee_df["Open"].astype("float64")
nee_df = nee_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

pg_df["pg_open"] = pg_df["Open"].astype("float64")
pg_df = pg_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

xom_df["xom_open"] = xom_df["Open"].astype("float64")
xom_df = xom_df.drop(columns = ["Open", "High", "Low", "Close", "Adj Close", "Volume"])

open_1 = pd.merge(aapl_df, amzn_df, on="Date", how="left")
open_2 = pd.merge(open_1, brk_df, on="Date", how="left")
open_3 = pd.merge(open_2, hon_df, on="Date", how="left")
open_4 = pd.merge(open_3, jnj_df, on= "Date", how="left")
open_5 = pd.merge(open_4, lin_df, on= "Date", how="left")
open_6 = pd.merge(open_5, nee_df, on= "Date", how= "left")
open_7 = pd.merge(open_6, pg_df, on= "Date", how= "left")
open_df = pd.merge(open_7, xom_df, on = "Date", how= "left")
open_df.to_csv("open_df.csv")

open_df["Date"] = open_df["Date"].astype(str)
open_df["Date"] = pd.to_datetime(open_df['Date'])
open_df['Year'] = open_df['Date'].dt.year
open_df['Month'] = open_df['Date'].dt.month
open_df['Day'] = open_df['Date'].dt.day


# open_df = open_df.loc[(open_df[""])]

# for i in len(range(open_df["Date"]):
#              index.append(i)

             
# for i in index: 
#     if 

# new_df = open_df.loc(open_df["Date"[6]] != ["Date"[6]])


# # new_list = []
# # for i, row in open_df.iterrows():
# #     if row["Date"][6] != row["Date"][5]:
# #         new_list.append(row)
# # print(new_list)
# new_df

# plt.plot(open_df["Date"], open_df["aapl_open"])
# plt.plot(open_df["Date"], open_df["amzn_open"])
# plt.plot(open_df["Date"], open_df["brk_open"])
# plt.plot(open_df["Date"], open_df["hon_open"])
# plt.plot(open_df["Date"], open_df["jnj_open"])
# plt.plot(open_df["Date"], open_df["lin_open"])
# plt.plot(open_df["Date"], open_df["nee_open"])
# plt.plot(open_df["Date"], open_df["pg_open"])
# plt.plot(open_df["Date"], open_df["xom_open"])
# plt.show
```

```python
#  Doing For Loops to find Annual % returns
# APPLE DATA 
apple_year=[]
apple_year_change=[]
row=0
open_price = 0
for index, row in filtered_df.iterrows():
    if index != 0:
        # do calculations
        price_change = (float(str(row[2])) - float(str(open_price)))/ float(str(open_price))
        print(index)
        print(row)
        open_price = row[2]
        open_year=row[11]
        apple_year_change.append(price_change)
        apple_year.append(open_year)
    else:
        # set first open price here
        open_price=row[2]
        continue
print(apple_year_change)
```

```python
mean_change.plot.bar()
plt.grid()
plt.axhline(label= "S&P500 growth", y = 6.75, color = 'r', linestyle = '--', lw=2)
plt.ylabel('Average % Change', fontsize=13)
plt.xlabel('Company', fontsize=13)
plt.title('Top Company Per Sector Average Annual % Increase From 1999-2020', fontsize=16)
plt.text(9,7, "--- = Average S&P500 % Return", fontsize=12, color="red", bbox=dict(facecolor='white', alpha=0.5))
plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize=13,
)
plt.savefig("Top Company Average Yearly Increase")
```
