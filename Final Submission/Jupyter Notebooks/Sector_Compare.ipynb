```python
# Import dependencies

# This will allow us to create file paths across operating systems
import os

# Module for reading CSV files
import csv

# This will allow us to use pandas
import pandas as pd

# Import Numpy for calculations and matplotlib for charting
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import CSV files for each industry ETF

#Materials
csvxlb_path = os.path.join('Yahoo', 'XLB.csv' )
csvxlb = pd.read_csv(csvxlb_path)

#Communications
csvxlc_path = os.path.join('Yahoo', 'XLC_2018.csv' )
csvxlc = pd.read_csv(csvxlc_path)

#Energy
csvxle_path = os.path.join('Yahoo', 'XLE.csv' )
csvxle = pd.read_csv(csvxle_path)

#Financial
csvxlf_path = os.path.join('Yahoo', 'XLF.csv' )
csvxlf = pd.read_csv(csvxlf_path)

#Industrial
csvxli_path = os.path.join('Yahoo', 'XLI.csv' )
csvxli = pd.read_csv(csvxli_path)

#Technology
csvxlk_path = os.path.join('Yahoo', 'XLK.csv' )
csvxlk = pd.read_csv(csvxlk_path)

#Consumer Staples
csvxlp_path = os.path.join('Yahoo', 'XLP.csv' )
csvxlp = pd.read_csv(csvxlp_path)

#Real Estate
csvxlre_path = os.path.join('Yahoo', 'XLRE_2015.csv' )
csvxlre = pd.read_csv(csvxlre_path)

#Utilities
csvxlu_path = os.path.join('Yahoo', 'XLU.csv' )
csvxlu = pd.read_csv(csvxlu_path)

#Health Care
csvxlv_path = os.path.join('Yahoo', 'XLV.csv' )
csvxlv = pd.read_csv(csvxlv_path)

#Consumer Discretionary
csvxly_path = os.path.join('Yahoo', 'XLY.csv' )
csvxly = pd.read_csv(csvxly_path)

#S&P 500
csvspy_path = os.path.join('Yahoo', 'SPY.csv' )
csvspy = pd.read_csv(csvspy_path)
```

```python
#Remove extra columns so we just have Date, Open, and Close for each df

xlb_columns = csvxlb[['Date', 'Open', 'Close']]
xlb = xlb_columns.rename(columns = {'Open' : "XLB Open" , 'Close' : 'XLB Close'})    

xlc_columns = csvxlc[['Date', 'Open', 'Close']]
xlc = xlc_columns.rename(columns = {'Open' : "XLC Open" , 'Close' : 'XLC Close'})    

xle_columns = csvxle[['Date', 'Open', 'Close']]
xle = xle_columns.rename(columns = {'Open' : "XLE Open" , 'Close' : 'XLE Close'})    

xlf_columns = csvxlf[['Date', 'Open', 'Close']]
xlf = xlf_columns.rename(columns = {'Open' : "XLF Open" , 'Close' : 'XLF Close'})    

xli_columns = csvxli[['Date', 'Open', 'Close']]
xli= xli_columns.rename(columns = {'Open' : "XLI Open" , 'Close' : 'XLI Close'})    

xlk_columns = csvxlk[['Date', 'Open', 'Close']]
xlk = xlk_columns.rename(columns = {'Open' : "XLK Open" , 'Close' : 'XLK Close'})    

xlp_columns = csvxlp[['Date', 'Open', 'Close']]
xlp = xlp_columns.rename(columns = {'Open' : "XLP Open" , 'Close' : 'XLP Close'})    

xlre_columns = csvxlre[['Date', 'Open', 'Close']]
xlre = xlre_columns.rename(columns = {'Open' : "XLRE Open" , 'Close' : 'XLRE Close'})    

xlu_columns = csvxlu[['Date', 'Open', 'Close']]
xlu = xlu_columns.rename(columns = {'Open' : "XLU Open" , 'Close' : 'XLU Close'})  

xlv_columns = csvxlv[['Date', 'Open', 'Close']]
xlv = xlv_columns.rename(columns = {'Open' : "XLV Open" , 'Close' : 'XLV Close'})    

xly_columns = csvxly[['Date', 'Open', 'Close']]
xly = xly_columns.rename(columns = {'Open' : "XLY Open" , 'Close' : 'XLY Close'})   

spy_columns = csvspy[['Date', 'Open', 'Close']]
spy = spy_columns.rename(columns = {'Open' : "SPY Open" , 'Close' : 'SPY Close'})  
```

```python
#Merge all CSV 1998 files on monthly date

b_and_e = pd.merge(xlb, xle, on="Date", how="left")
f = pd.merge(b_and_e, xlf, on="Date", how="left")
i = pd.merge(f, xli, on="Date", how="left")
k = pd.merge(i, xlk, on="Date", how="left")
p = pd.merge(k, xlp, on="Date", how="left")
u = pd.merge(p, xlu, on="Date", how="left")
v = pd.merge(u, xlv, on="Date", how="left")
y = pd.merge(v, xly, on="Date", how="left")
df_1999 = pd.merge(y, spy, on="Date", how="left")
```

```python
#drop any NAN
df_1999 = df_1999.dropna()
```

```python
df_1999
```

```python
# Convert date column to datetime format & add year, month, day columns

df_1999['Date']=pd.to_datetime(df_1999['Date'], format='%Y-%m-%d')
df_1999['Year']= df_1999['Date'].dt.year
df_1999['Month'] = df_1999['Date'].dt.month
df_1999['Day'] = df_1999['Date'].dt.day
```

```python
df_1999
```

```python
# Only pull January months

df_1999=df_1999.loc[df_1999['Month'] == 1]
df_1999=df_1999.reset_index(drop=True)
df_1999
```

```python
# df_1999 = df_1999.drop(index = 22)
# df_1999
```

```python
#Drop Close Prices and Month/Day/Year
open_only=df_1999.drop(columns=['XLB Close','XLE Close', 'XLF Close', 'XLI Close', 'XLK Close', 'XLP Close', 'XLU Close', 'XLV Close','XLY Close' , 'SPY Close', 'Year', 'Month' , 'Day'],axis=1)
open_only.head()
```

```python
#Rename tickers to industries

sectors = open_only.rename(columns={
                    "XLB Open":"Materials", 
                    "XLE Open":"Energy",
                    "XLF Open":"Financial", 
                    "XLI Open":"Industrial",
                    "XLK Open":"Technology", 
                    "XLP Open":"Consumer Staples",
                    "XLV Open":"Health Care", 
                    "XLY Open":"Consumer Discretionary",
                    "XLU Open":"Utilities",
                    "SPY Open":"S&P 500",
})
sectors
```

```python
#set date as index

sectors.set_index('Date',inplace=True)
sectors
```

```python
#Convert monthly data to the average at the end of each year
# yearly_average_1999 = sectors.resample('Y').mean()
# yearly_average_1999.head()
```

```python
# Year over year change %

pct_change = sectors.pct_change()*100
pct_change
```

```python
#Average the values over the time period

average_pct = pct_change.mean()
average_pct
```

```python
#Round the data

rounded_avg = average_pct.round(2)
rounded_avg
```

```python
#plot the data

rounded_avg.plot.bar()
plt.axhline(y = 6.71, color = 'r', linestyle = '--', lw=2)
plt.ylabel('Average Annual % Return', fontsize=13)
plt.xlabel('Sectors', fontsize=13)
plt.title('S&P 500 Sectors Annual Avg. % Performance From 1999-2020', fontsize=16)

plt.xticks(
    rotation=45,
    horizontalalignment='right',
    fontweight='light',
    fontsize=13,
)
```
